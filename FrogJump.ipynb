```python
#青蛙跳台阶问题
import numpy as np
#用数组记录步数，减低时间成本，用空间换时间
def FrogJump(n, recod):
    if recod[n-1] == 0:
        if n == 1:
            recod[n-1] = 1
            return 1, recod

        if n == 2:
            recod[n-1] = 2
            return 2, recod
    
        a, recod = FrogJump(n-1,recod)

        b, recod = FrogJump(n-2,recod)

        recod[n-1] = a + b
    return recod[n-1], recod
```

```python
n = 1000   #若不用数组记录有些节点会重复查询，运行过慢
rec = np.zeros(n)
num, rec = FrogJump(n,rec)
num
```
